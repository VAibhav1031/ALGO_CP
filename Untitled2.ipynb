```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graphs
output = np.arange(1, 21)  # Output levels from 1 to 20
fixed_cost = 100  # Fixed cost remains constant
variable_cost = 10 * output  # Variable cost increases linearly
total_cost = fixed_cost + variable_cost  # Total cost is the sum of fixed and variable costs

average_fixed_cost = fixed_cost / output  # AFC decreases as output increases
average_variable_cost = variable_cost / output  # AVC is constant since variable cost is linear
average_total_cost = total_cost / output  # ATC is the sum of AFC and AVC
marginal_cost = np.gradient(total_cost)  # Marginal cost is the rate of change of total cost

# Plotting graphs
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Total Cost, Fixed Cost, and Variable Cost
axes[0, 0].plot(output, fixed_cost * np.ones_like(output), label="Fixed Cost (TFC)", linestyle="--")
axes[0, 0].plot(output, variable_cost, label="Variable Cost (TVC)", linestyle="-.")
axes[0, 0].plot(output, total_cost, label="Total Cost (TC)", color="black")
axes[0, 0].set_title("Total, Fixed, and Variable Costs")
axes[0, 0].set_xlabel("Output")
axes[0, 0].set_ylabel("Cost")
axes[0, 0].legend()
axes[0, 0].grid(True)

# Average Fixed Cost (AFC), Average Variable Cost (AVC), and Average Total Cost (ATC)
axes[0, 1].plot(output, average_fixed_cost, label="Average Fixed Cost (AFC)", linestyle="--")
axes[0, 1].plot(output, average_variable_cost, label="Average Variable Cost (AVC)", linestyle="-.")
axes[0, 1].plot(output, average_total_cost, label="Average Total Cost (ATC)", color="black")
axes[0, 1].set_title("Average Costs")
axes[0, 1].set_xlabel("Output")
axes[0, 1].set_ylabel("Cost per Unit")
axes[0, 1].legend()
axes[0, 1].grid(True)

# Marginal Cost (MC)
axes[1, 0].plot(output, marginal_cost, label="Marginal Cost (MC)", color="blue")
axes[1, 0].set_title("Marginal Cost")
axes[1, 0].set_xlabel("Output")
axes[1, 0].set_ylabel("Marginal Cost")
axes[1, 0].legend()
axes[1, 0].grid(True)

# Combined Average and Marginal Cost
axes[1, 1].plot(output, average_total_cost, label="Average Total Cost (ATC)", color="green")
axes[1, 1].plot(output, average_variable_cost, label="Average Variable Cost (AVC)", linestyle="--")
axes[1, 1].plot(output, marginal_cost, label="Marginal Cost (MC)", color="blue")
axes[1, 1].set_title("Combined Cost Curves")
axes[1, 1].set_xlabel("Output")
axes[1, 1].set_ylabel("Cost")
axes[1, 1].legend()
axes[1, 1].grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
